# Print out experiment metadata

```python
import os
import pandas as pd
from pprint import pprint

import cam_api as api
```

```python
CAM_DIR_PATH = '/Volumes/Brain2015/CAM/'
```

```python
expt_ids = os.walk(CAM_DIR_PATH).next()[1]
expt_paths = [os.path.join(CAM_DIR_PATH, expt_id) for expt_id in expt_ids]
nwb_paths = [os.path.join(expt_path, expt_id + '.nwb') for expt_path, expt_id in zip(expt_paths, expt_ids)]

meta_data = pd.DataFrame(index=range(len(nwb_paths)), columns=api.getMetaData(nwb_paths[0]).keys() + ['path'])
for ctr, nwb_path in enumerate(nwb_paths):
    meta_data.loc[ctr] = pd.Series(api.getMetaData(nwb_path))
    meta_data['path'][ctr] = nwb_path
results = meta_data.sort(['HVA', 'depth', 'Cre'])[['HVA', 'depth', 'Cre', 'path']]
```
