```python
import os
import pandas as pd
from pprint import pprint

import api
import metrics
```

```python
CAM_DIR_PATH = '/Volumes/Brain2015/CAM/'
DETREND_WINDOW = 180
NEURONS_PER_STIM = 10
```

```python
nwb_paths = api.get_nwb_paths(CAM_DIR_PATH)

columns = api.get_meta_data(nwb_paths[0]).keys() + ['path', 'variability']
meta_data = pd.DataFrame(index=range(len(nwb_paths)), columns=columns)

for ctr, nwb_path in enumerate(nwb_paths):
    meta_data.loc[ctr] = pd.Series(api.get_meta_data(nwb_path))
    meta_data['path'][ctr] = nwb_path
    
    # calculate experiment variability
    var = metrics.variability_stim_conditioned(
        nwb_path, DETREND_WINDOW, NEURONS_PER_STIM
    )['variability'].mean()
    meta_data['variability'][ctr] = var
    
results = meta_data.sort(['variability'])
results = results[['variability', 'HVA', 'depth', 'Cre', 'path']]
```
